```python
import matplotlib.pyplot as plt

def dda_line(x0, y0, x1, y1):
    dx = x1 - x0
    dy = y1 - y0
    steps = max(abs(dx), abs(dy))  # Liczba krokw
    
    x_incr = dx / steps
    y_incr = dy / steps

    x, y = x0, y0
    pixels = []

    for _ in range(int(steps) + 1):  # Iteracja o jeden wicej, aby uwzgldni ostatni punkt
        pixels.append((round(x), round(y)))  # Zaokrglone wsprzdne
        x += x_incr
        y += y_incr

    return pixels

# Testowanie funkcji
pixels = dda_line(200, 180, 180, 160)

# Rysowanie wykresu
x_coords, y_coords = zip(*pixels)
plt.figure(figsize=(6, 6))
plt.scatter(x_coords, y_coords, c='black', marker='s', s=50)
plt.plot(x_coords, y_coords, linestyle='dashed', color='gray', alpha=0.5)

# Ustawienia osi
plt.gca().invert_yaxis()  
plt.xticks(range(min(x_coords)-1, max(x_coords)+2, 2))
plt.yticks(range(min(y_coords)-1, max(y_coords)+2))
plt.grid(True, linestyle="--", linewidth=0.5)

# Opis wykresu
plt.title("Poprawiona wizualizacja DDA")
plt.xlabel("X")
plt.ylabel("Y")

# Wywietlenie wykresu
plt.show()

# Wypisanie pikseli
print(pixels)
```
